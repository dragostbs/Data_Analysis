## Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
var_norm=norm()
```

#### Ex 1.
$P(Z\le 1.25)$

```python
var_norm.cdf(1.25)
```

```python
x=np.linspace(-4,4,100)
plt.plot(x, var_norm.pdf(x))
plt.fill_between(x, var_norm.pdf(x), where= x<=1.25)
plt.title(f"$P(Z< 1.25)$={var_norm.cdf(1.25)}")
plt.show()
```

#### Ex 2.
$P(Z<-1.25)$

```python
var_norm.cdf(-1.25)
```

```python
x=np.linspace(-4,4,100)
plt.plot(x, var_norm.pdf(x))
plt.fill_between(x, var_norm.pdf(x), where= x<=-1.25)
plt.title(f"$P(Z< -1.25)$={var_norm.cdf(-1.25)}")
plt.show()
```

#### Ex 3.
$P(Z>0.4)$

```python
1-var_norm.cdf(0.4)
```

```python
x=np.linspace(-4,4,100)
plt.plot(x, var_norm.pdf(x))
plt.fill_between(x, var_norm.pdf(x), where= x>=0.4)
plt.title(f"$P(Z> 0.4)$={1-var_norm.cdf(0.4)}")
plt.show()
```

#### Ex 4.
$P(0.2<Z<0.8)$

```python
var_norm.cdf(0.8)-var_norm.cdf(0.2)
```

```python
x=np.linspace(-4,4,100)
plt.plot(x, var_norm.pdf(x))
plt.fill_between(x, var_norm.pdf(x), where= np.logical_and(x>0.2, x<0.8))
plt.title(f"$P(0.2<Z<0.8)$={var_norm.cdf(0.8)-var_norm.cdf(0.2)}")
plt.show()
```

### The time required to cover the distance from point A to point B can be described using a normal distribution with an average of 45 minutes and a standard deviation of 6 minutes.
- what percentage of trips take less than 35 minutes?
- what percentage of trips take more than 60 minutes?
- How long should a person spend on this trip so as not to be late in 70% of cases?

#### Normal distribution Graph representation

```python
var_norm=norm(45, 6)
Y=np.linspace(20, 70, 100)
plt.plot(Y, var_norm.pdf(Y))
plt.show()
```

#### Ex 1.

```python
var_norm.cdf(35)
```

```python
Y=np.linspace(20, 70, 100)
plt.plot(Y, var_norm.pdf(Y))
plt.fill_between(Y, var_norm.pdf(Y), where= Y<35)
plt.title("Percentage of trips lasting less than 35 minutes")
plt.show()
```

#### Ex 2.

```python
1-var_norm.cdf(60)
```

```python
Y=np.linspace(20, 70, 100)
plt.plot(Y, var_norm.pdf(Y))
plt.fill_between(Y, var_norm.pdf(Y), where= Y>60)
plt.title("Percentage of trips lasting over 60 minutes")
plt.show()
```
