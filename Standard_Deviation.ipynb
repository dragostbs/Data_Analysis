# Standard Deviation

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

#### Ex 1. The average hourly wage of farmers is 13.5 units. It is known that Ïƒ = 2.9 units. Calculate and interpret the standard error of the mean for a sample size of 30.

```python
mu=13.5
sigma=2.9
n=30

standardError=sigma/np.sqrt(n)
standardError
```

```python
x=np.linspace(mu-3.5*standardError, mu+3.5*standardError, 100)
plt.plot(x, norm.pdf(x, loc=mu, scale=standardError))
plt.title("average sample size distribution of 30")
plt.show()
```

#### Ex 2. We assume that we have a sample that contains the yield for the last 30 months of a company X. The average of these values â€‹â€‹is 2% and the average square deviation of the sample is 20%. Calculate and interpret the standard error of the sample mean.

```python
# s - mean square deviation for the sample !

mu=2
n=30
s=20

standardError=s/np.sqrt(n)
standardError
```

```python
x=np.linspace(mu-3.5*standardError, mu+3.5*standardError, 100)
plt.plot(x, norm.pdf(x, loc=mu, scale=standardError))
plt.title("average sample size distribution of 30")
plt.show()
```
